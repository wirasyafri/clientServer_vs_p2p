```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
N, F, us, dmin, ui = sp.symbols('N F U_s d_min U_i')
```

```python
display(dmin)
```

```python
eq = N*F/us
eq2 = F/dmin
display(eq)
display(eq2)
```

```python
eq_lmbd = sp.lambdify((N,F,us),eq)
```

```python
print(eq_lmbd(5,2,10))
```

```python
nilai_N = np.linspace(1,100,100) 
print(nilai_N)
```

```python
data_n = []
for n in nilai_N:
    data_n.append(eq_lmbd(n,2,10))
print(data_n)
```

```python
plt.plot(nilai_N, data_n)
plt.xlabel('N')
plt.ylabel('$N/(FU_s)$')
```

```python
eq3 = N*F/(us + (N*ui))
display(eq3)
```

```python
eq3_lmbd = sp.lambdify((N,F,us,ui),eq3)
```

```python
print(eq3_lmbd(1,2,10,1))
```

```python
data_p = []
for p in nilai_N:
    data_p.append(eq3_lmbd(p,2,10,1))
print(data_p)
```

```python
plt.plot(nilai_N, data_p)
plt.xlabel('N')
plt.ylabel('$NF/(U_s+NU_i)$')
```

```python
plt.rcParams["figure.autolayout"] = True
line1, = plt.plot(nilai_N, data_p, label="peer to perr")
line2, = plt.plot(nilai_N, data_n, label="client server")
leg = plt.legend(loc='upper center')
plt.xlabel('N')
plt.ylabel('$NF/(Minimum Distribution Time)$')
```

```python
randomUserUpload = []
def createRandUserUp(user):
    sumUserUp=0
    for i in range(0,user):
        n = random.uniform(1,2)
        randomUserUpload.append(n)
        sumUserUp += n
    return sumUserUp

createRandUserUp(1)
```

```python
sum_ui = sp.symbols('Σ_ui')
eq4= N*F/(us + sum_ui)
display(eq4)

eq4_lmbd = sp.lambdify((N,F,us,sum_ui),eq4)
# print(eq4_lmbd(1,2,10,createRandUserUp(1)))

data_q = []
for q in nilai_N:
    data_q.append(eq4_lmbd(q,2,10,createRandUserUp(int(q))))
print(data_q)

```

```python
plt.rcParams["figure.autolayout"] = True
line1, = plt.plot(nilai_N, data_q, label="peer to perr w/ random upload rate user")
line2, = plt.plot(nilai_N, data_n, label="client server")
leg = plt.legend(loc='upper center')
plt.xlabel('N')
plt.ylabel('$NF/(Minimum Distribution Time)$')
```
